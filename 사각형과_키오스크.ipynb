```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')
            print(self.rect())  # 직사각형의 넓이를 계산하고 출력합니다.

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
            print(self.par())  # 평행사변형의 넓이를 계산하고 출력합니다.

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
            print(self.trape())  # 사다리꼴의 넓이를 계산하고 출력합니다.

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

  # 평행사변형의 넓이
    def par(self):
        base, height = map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))
        area = base * height
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

    # 사다리꼴의 넓이
    def trape(self):
        base1, base2, height = map(int, input('윗변, 아랫변, 높이를 입력하세요. 예시 : 윗변,아랫변,높이\n >>>').split(','))
        area = (base1 + base2) * height / 2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result

a = Square()
```

```python
class Kiosk:
    def __init__(self):
        # 메뉴와 가격 리스트 초기화
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트
        self.total_price = 0  # 총 금액

    # 메뉴와 가격을 출력하는 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        # 첫 번째 음료 주문
        n = 0
        while n < 1 or n > len(self.menu):
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력 (1,2,3,4,5,6)

            if 1 <= n <= len(self.menu):
                self.price_sum = self.price[n - 1]  # 선택 음료의 가격

                # 음료 온도 물어보기
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = "HOT"
                    elif t == 2:
                        temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                # 첫 번째 음료를 주문 리스트와 가격 리스트에 추가
                self.order_menu.append(f"{temp} {self.menu[n - 1]}")
                self.order_price.append(self.price_sum)
                self.total_price += self.price_sum

                # 첫 번째 음료 주문 결과 출력
                print(f'주문한 음료: {self.menu[n - 1]} ({temp}) : {self.price_sum} 원\n합계 : {self.total_price} 원')

            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 추가 주문 또는 지불
        while True:
            print()
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불

            if n > 0 and n <= len(self.menu):
                # 추가 음료 주문
                self.price_sum = self.price[n - 1]  # 선택 음료의 가격

                # 추가 음료 온도 물어보기
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = "HOT"
                    elif t == 2:
                        temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                # 추가 음료를 주문 리스트와 가격 리스트에 추가
                self.order_menu.append(f"{temp} {self.menu[n - 1]}")
                self.order_price.append(self.price_sum)
                self.total_price += self.price_sum

                # 추가 주문 결과 출력
                print(f'추가 주문 음료: {self.menu[n - 1]} ({temp}) : {self.price_sum} 원\n합계 : {self.total_price} 원')

            elif n == 0:
                # 지불을 선택하면 주문 완료 메시지 출력
                print("주문이 완료되었습니다.")
                self.table()  # 주문표 출력
                break  # 반복 종료

            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    # 주문 내역을 표 형식으로 출력하는 메서드
    def table(self):

        print('⟝' + '-' * 30 + '⟞')
        for _ in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명과 가격을 출력
        for i in range(len(self.order_menu)):
            print(f'{self.order_menu[i]:<30} : {self.order_price[i]:>4} 원')

        # 총 금액을 출력
        print(f'합계 금액 : {self.total_price:>4} 원')

        for _ in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')

    # 지불 메서드
    def pay(self):
        payment_method = int(input("지불 방법을 선택하세요: 현금은 1, 카드 결제는 2 : "))

        if payment_method == 1:
            print("직원을 호출하겠습니다.")
        elif payment_method == 2:
            print("IC칩 방향에 맞게 카드를 꽂아주세요.")
        else:
            print("잘못된 입력입니다. 지불 방법을 다시 선택해주세요.")

# Kiosk 객체 생성 및 메뉴 출력
kiosk = Kiosk()
kiosk.menu_print()

# 음료 선택 및 추가 주문
kiosk.menu_select()
```
